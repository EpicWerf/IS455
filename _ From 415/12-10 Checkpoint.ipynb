```
# Jackson van der Werf
# This program will scrape a list of cat breeds from a Wikipedia page and download it as a .csv file

# import BeautifulSoup and Requests
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_cat_breeds'

req = requests.get(url)
req
```

```
# parse webpage and create a BeautifulSoup object
html = req.text
soup = BeautifulSoup(html, "html.parser")
soup.table
```

```
# import pandas, scrape table on the webpage, write table to a pandas df
import pandas as pd
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
df
```

```
# download the df created by pandas to a .csv file
from google.colab import files

df.to_csv('df.csv')
files.download('df.csv')
```

```
from bs4 import BeautifulSoup
import requests

url = "http://books.toscrape.com/"
```

```
# request the HTML from the url specified before
req = requests.get(url)

# the following code is optional
# it prints out if the request was successful
req
```

```
html = req.text
html
```

```
soup = BeautifulSoup(html, 'html.parser')
soup
```

```
soup.title.text
print(soup.title.text.replace('\n', '').strip())
```

```
book_urls = []
for i in range(1, 6):
  url = f'http://books.toscrape.com/catalogue/page-{i}.html'
  req = requests.get(url)
  soup = BeautifulSoup(req.text, 'html.parser')
  for h3 in soup.find_all('h3'):
    book_urls.append(h3.a['href'])


for url in book_urls:
  print(url)
```

```
import pandas as pd

df = pd.DataFrame(columns=['UPC', 'title', 'price', 'rating', 'available', 'desc'])

for url in book_urls:
  req = requests.get(f'http://books.toscrape.com/catalogue/{url}')
  soup = BeautifulSoup(req.text, 'html.parser')
  upc = soup.table.tr.td.text
  print(upc)

  break
```

```python
import pandas as pd, numpy as np, seaborn as sns, statsmodels.api as sm

df = pd.read_csv('https://www.ishelp.info/data/insurance.csv')
df.head()
```

```python
label = df['charges']
# features = df[['age', 'children', 'bmi']].assign(const=1) #manual way. Auto way is below
features = df.select_dtypes(include=[np.number]).assign(const=1) #you have to force const (y-int) otherwise r2 is way too high
features.drop(['charges'], axis=1, inplace=True) #drop the label column

model = sm.OLS(label, features).fit()
print(model.summary())
```

```python
# new version of the label to reduce skew
features = np.log(features)

model = sm.OLS(label, features).fit()
print(model.summary())
```

```python
df['predictions'] = model.fittedvalues
df.head()
```

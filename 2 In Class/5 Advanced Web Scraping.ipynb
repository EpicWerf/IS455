```python
#visit the grades page
url_grades = 'https://app.myeducator.com/course/activities/1702c/'
req_grades = session_requests.get(url_grades)
req_grades.text
```

```python
#create a soup object out of the grades page
soup = BeautifulSoup(req_grades.text, 'html.parser')

#find all script tags
scripts = soup.find_all('script')

for script in scripts:
    print(script.text)
```

```python
#use regex to find script tag matching what we want
import re
script = soup.find('script', text=re.compile(r'startCourseActivities'))

#find the text that starts the dictionary portion
start_text = '{"courseData'
start_position = script.text.index(start_text)
text = script.text[start_position:-12]

#get rid of type issues that don't let you cast to a dictionary
text = text.replace('true', 'True')
text = text.replace('false', 'False')
text = text.replace('null', '')
```

```python
import ast

#print out as a dictionary
data = ast.literal_eval(text)
data['groups'][0]

for group in data['groups']:
    for activity in group['activities']:
        for item in activity[3]:
            score = item[1].split(' / ')
            numerator += float(score[0])
            denominator += float(score[1])

print(f'Current grade:\t {numerator} / {denominator}')
print(f'Current percentage:\t{round((numerator/denominator)*100, 2)}%')

# scores = []
# for group in data['groups']:
#     for activity in group['activities']:
#         for item in activity[3]:
#             scores.append(item[2])
#             print(item[2])
# scores
```

```python
#Get the data and turn it into a clean JSON object
import requests, json

url = "https://www.balldontlie.io/api/v1/players"

response = requests.get(url)
json_data = json.loads(response.text)
json_clean = json.dumps(json_data, indent=2)

print(json_clean)
```

```python
#put the json object into a dataframe
import pandas as pd

# for k, v in json_data.items():
#     print(k, v)

for player in json_data['data']:
    print(player)
```

```python
#Get column headers
# print(json_data['data'][0].keys())

import time

#put the keys as the column headers
df_players = pd.DataFrame(columns=json_data['data'][0].keys())
df_players.set_index('id', inplace=True)
df_teams = pd.DataFrame(columns=json_data['data'][0]['team'].keys())
df_teams.set_index('id', inplace=True)

#add all the players to a dataframe. Make sure to get all pages
has_next = True
page=1
total_pages=json_data['meta']['total_pages']
url = f"https://www.balldontlie.io/api/v1/players?page={page}"

while has_next: 
    response = requests.get(url)
    json_data = json.loads(response.text)

    for player in json_data['data']:
        # print(player.keys())
        df_players.loc[player['id']] = player
    print(f'scraping page {page}')
    page += 1
    if page > total_pages:
        has_next = False
    time.sleep(1)

#add all the teams to a dataframe
# for team in json_data['data']:
#     df_teams.loc[team['id']] = team

df_teams
```

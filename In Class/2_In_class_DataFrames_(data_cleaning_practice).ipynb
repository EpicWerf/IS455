```python
import pandas as pd
pd.set_option('display.max_rows', 1000)

df = pd.read_csv('Student_Enrollment_Sample.csv', low_memory=False)
print(df.shape)
print(df.isna().sum())
df.head()
```

```python
df.EXPECTED_START_DATE.isnull().sum()
```

```python
df_no_missing = df.copy()

for col in df_no_missing.columns:
    if df_no_missing[col].isnull().sum() > 10000:
        df_no_missing.drop(columns=col, inplace=True)

df_no_missing.dropna(inplace=True)
print(df_no_missing.isna().sum())

df_no_missing.shape
```

```python
df = df_no_missing.copy()

# Remove rows with in credits attempted that have 0 credits attempted
df = df[df['CREDITS_ATTEMPTED'] > 0 ]

df['Completion_Rate'] = df.CREDITS_EARNED / df.CREDITS_ATTEMPTED

df
```

```python
print(df.STATUS_DESCRIPTION.value_counts())
df['LABEL'] = 0
df
```

```python
df.loc[df['STATUS_DESCRIPTION'] == 'Active', 'LABEL'] = 2
df.loc[df['STATUS_DESCRIPTION'] == 'Active-Pre Grad', 'LABEL'] = 2
df.loc[df['STATUS_DESCRIPTION'] == 'Graduate', 'LABEL'] = 2
df.loc[df['STATUS_DESCRIPTION'] == 'Transfer From Other Program', 'LABEL'] = 2
df.loc[df['STATUS_DESCRIPTION'] == 'Transfer To Other Program', 'LABEL'] = 2
df
```

```python
# Itterate through each row

for item, row in enumerate(df.itertuples()) :
    if item < 6:
        print(f'{item}')
        print(row[1])
df
```

```python
# get the script tag containing our grades using regex
import re

page = session_requests.get("https://app.myeducator.com/course/activities/1702c/")
soup = BeautifulSoup(page.text, 'html.parser')
script = soup.findAll("script", text=re.compile(r'MyEducator.startCourseActivities'))
script
```

```python
import ast, numpy as np

item = script[0]         # Get the first item from the list (i.e. remove brackets)
text = item.text         # Get the text from the tag (i.e. remove script tags)
s = 'courseData'         # This is the first dictionary item
start = text.index(s)    # Find where this text beings
trim = text[start - 2:]  # Remove unnecessary string from beginning
trim = trim[:-12]        # Remove unnecessary string from end
trim = trim.replace('true', 'True')  # Change true value to Python-compatible version
trim = trim.replace('null', '')      # Update null values to an empty string
data = ast.literal_eval(trim)        # Cast the remaining text to a Python dictionary
data
```

```python
#Now that we have a workable dictionary, let's iterate through the results and pull out the grades.
# The typical way to iterate over dictionaries is using the .items() method which extracts a list of key/value pairs 
# (although you can use .keys() and .values() as well to get only one of them):
for key, value in data.items():
    print(key, value)
```

```python
#We can see from this that we want to get the list from the "groups" value. 
# Let's iterate again just over the list elements from "groups":
for item in data['groups']:
    print(item)
```

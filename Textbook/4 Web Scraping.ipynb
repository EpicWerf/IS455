## **Storing Scraped Data in a DataFrame**

### Collect URLs for Product Details

```
from bs4 import BeautifulSoup
import requests

url = 'http://books.toscrape.com/catalogue/page-'
books = []
for i in range(1, 6):
  u = f'{url}{i}.html'
  req = requests.get(u)
  html = req.text
  soup = BeautifulSoup(html, 'html.parser')
  for book_url in soup.find_all('a', href=True):    # href=True parameter allows us to only select <a> tags that have the href attribute
    books.append(book_url['href'])                  # Notice we can reference HTML attributes like an index using the attribute name

books
```

```
url = 'http://books.toscrape.com/catalogue/page-1.html'
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
soup
```

```
url = 'http://books.toscrape.com/catalogue/page-'
books = []
for i in range(1, 6):
  u = f'{url}{i}.html'
  req = requests.get(u)
  html = req.text
  soup = BeautifulSoup(html, 'html.parser')
  for h3 in soup.find_all('h3'):
    books.append(h3.a['href'])    # Notice that now we reference the 'href' attribute of the 'a' tag within the 'h3' tag 

books
```

```
books = []
more_books = True
i = 1
while more_books:
  req = requests.get(f'http://books.toscrape.com/catalogue/page-{i}.html')
  print(f'{req.status_code}, ', end="")  # <- We are printing this only so that you can see the loop execute; comment it out if you don't want it
  i += 1                          # increment the page

  if req.status_code != 200:      # If a 200 response is not received, then break
    break

  soup = BeautifulSoup(req.text, 'html.parser')
  for h3 in soup.find_all('h3'):
    books.append(h3.a['href'])    # Notice that now we reference the 'href' attribute of the 'a' tag within the 'h3' tag 

print('\n\n', books)              # <- Run this line if you want to verify that all of the book URLs were stored
```

### Cleaning Function

```
# Define the function:
def scrape_book_details(html):
  from bs4 import BeautifulSoup
  from word2number import w2n

  soup = BeautifulSoup(html, 'html.parser')
  title = soup.find("div", {"class": "col-sm-6 product_main"}).h1.text
  stock = soup.find("p", {"class": "instock availability"}).text.replace("\n", "").strip()
  rating = w2n.word_to_num(soup.find("p", {"class": "instock availability"}).next_sibling.next_sibling.get('class')[1])
  price = float(soup.find("p", {"class": "price_color"}).text[2:])
  desc = soup.find("div", {"id": "product_description"}).next_sibling.next_sibling.text.replace("\n", "")
  upc = soup.table.td.text

  return [upc, title, stock, rating, price, desc]

# Test the function:
req = requests.get('http://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html')
scrape_book_details(req.text)
```
